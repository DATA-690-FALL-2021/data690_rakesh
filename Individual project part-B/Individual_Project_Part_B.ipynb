```python
import pandas as pd
import plotly
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("/content/wdi_data.csv")
df.head()
```

```python
df.drop(columns=["Unnamed: 0"],inplace=True)
df
```

```python
df_ind = df[df["Country Name"]=='India']
df_ind.head()
```

```python
df_uk = df[df["Country Name"]=='United Kingdom']
df_uk.head()
```

```python
df_aus = df[df["Country Name"]=='Australia']
df_aus.head()
```

```python
df_Lanka = df[df["Country Name"]=='Sri Lanka']
df_Lanka.head()
```

```python
df_swit = df[df["Country Name"]=='Switzerland']
df_swit.head()
```

```python
frames=[df_ind,df_uk,df_aus,df_Lanka,df_swit]
result= pd.concat(frames)
result
```
