```python
#Reading the data
df = pd.read_csv("https://raw.githubusercontent.com/rakeshp1/data690_rakesh/main/Individual%20project%20part-B/wdi_data%20(4).csv")
df.head()
```

```python
df.info()
```

```python
#dropping unnecessary columns
df.drop(columns=["Unnamed: 0"],inplace=True)
df
```

# Choosing the India,United Kingdom,Australia,Switzerland and Sri Lanka Countries from the data

```python
df_ind = df[df["Country Name"]=='India']
df_ind.head()
```

```python
df_uk = df[df["Country Name"]=='United Kingdom']
df_uk.head()
```

```python
df_aus = df[df["Country Name"]=='Australia']
df_aus.head()
```

```python
df_Lanka = df[df["Country Name"]=='Sri Lanka']
df_Lanka.head()
```

```python
df_swit = df[df["Country Name"]=='Switzerland']
df_swit.head()
```

# Adding all the selected Countries in a separate Dataframe

```python
data_frame=[df_ind,df_uk,df_aus,df_Lanka,df_swit]
result= pd.concat(data_frame)
result
```

```python
df_2018 = result.query("Year == 2018") #choosing the selected year
df_2018.sample(10)
```

```python
df_2018 = df_2018.query("indicator in ['NE.EXP.GNFS.ZS','NV.IND.MANF.ZS','NV.IND.TOTL.ZS','BG.GSR.NFSV.GD.ZS','NV.AGR.TOTL.ZS','NY.GDP.PCAP.PP.CD']"	)
df_2018.head()
```

```python
df_pivot = df_2018.pivot(index =['Country Code','Country Name','Region','Income Group','Lending Type','Year'], columns='indicator', values='value').reset_index()
df_pivot.head()
```

```python
df_pivot
```

```python
df2 = df.query("indicator in ['NE.EXP.GNFS.ZS','NV.IND.MANF.ZS','NV.IND.TOTL.ZS','BG.GSR.NFSV.GD.ZS','NV.AGR.TOTL.ZS','NY.GDP.PCAP.PP.CD']")
df_pivot2 = df2.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns='indicator', values='value').reset_index()
df_pivot2.head()
```

# Exports of goods and services (% of GDP) contribution to Economy growth:

```python
fig = px.line(df_pivot2,
    x="Year",
    y="NE.EXP.GNFS.ZS",
    title="TOP 10 Countries by Exports of goods and services (% of GDP)",
    color='Country Name',
    labels={"Year":"Year","NE.EXP.GNFS.ZS":"Exports of goods and services (% of GDP)"},
    hover_name='Country Code')
fig.update_xaxes(type='category')
fig.show()

```

- From this graph we can observe that Switzerland is getting high economy from Exports of goods and services (% of GDP) than the remaining choosen countries
- And India is getting low economy from choosen countries.

# Agriculture, forestry, and fishing, value added (% of GDP) contribution to Economy growth:


```python
fig = px.bar(
    df_pivot,
    x="Country Name",
    y="NV.AGR.TOTL.ZS",
    title="TOP 10 Countries By 2018 Agriculture, forestry, and fishing, value added (% of GDP)",
    color='Country Name',
    labels={"NV.AGR.TOTL.ZS":"2018 Agriculture, forestry, and fishing, value added (% of GDP)"}
).update_xaxes(categoryorder="total descending")
fig.show()
```

- From this graph we can observe that India's (Agriculture, forestry, and fishing, value added (% of GDP)) is higher than remaining countries.
- United kingdom is getting less economy from (Agriculture, forestry, and fishing, value added (% of GDP)).

# Industry (including construction), value added (% of GDP) contribution to Economy growth

```python
fig = px.sunburst(
     df_pivot2, 
     path=['Region', 'Country Name'],
     values='NV.IND.TOTL.ZS',
     title='TOP 10 Countries by 2018 Industry (including construction), value added (% of GDP))',
     hover_name='Country Code',
 )

fig.show()
```

- From here we can observe that all the choosen countries are getting around 25% income from Industry (including construction), value added (% of GDP) except United Kingdom which is getting around 18% from industrial sector.

# Manufacturing, value added (% of GDP) contribution to Economy growth:

```python
fig = px.scatter(
    df_pivot.dropna(),
    x="NV.IND.MANF.ZS",
    y="NV.IND.MANF.ZS",
    title="TOP 10 Countries by 2018 Manufacturing, value added (% of GDP)",
    labels={"NV.IND.MANF.ZS":"2018 Manufacturing, value added (% of GDP)","NV.IND.MANF.ZS":"2018 Manufacturing, value added (% of GDP)"},
    color='Country Name',
    size='NV.IND.MANF.ZS',
    hover_name='Country Name',
    size_max=75,
    height=800,
)
fig.update_layout(
    showlegend=False
)
fig.update_traces(textposition ="top center")
fig.show()
```

- From this graph we can observe that Switzerland is getting high economy from Manufacturing, value added (% of GDP) than remaining choosen countries.

- And Australia is getting low economy growth from Manufacturing, value added (% of GDP).

# Trade in services (% of GDP) contribution to Economy growth

```python
fig = px.line(df_pivot2,
    x="Year",
    y="BG.GSR.NFSV.GD.ZS",
    title="TOP 10 Countries by Trade in Services (% of GDP)",
    color='Country Name',
    labels={"Year":"Year","BG.GSR.NFSV.GD.ZS":"Trade in services (% of GDP)"},
    hover_name='Country Code')

fig.show()

```

- If we observe this sector all the choosen countries Trade in services (% of GDP) contribution to the economy increased from 2010-2018, Switzerland is having highest percentage of contribution from Trade in services (% of GDP) compared to other countries.

# GDP per capita, PPP (current international)

```python
fig = px.pie(
    df_pivot,
    values='NY.GDP.PCAP.PP.CD', 
    names='Country Name',
    title='TOP 10 Countries by 2018 GDP per capita,PPP(Current International $)',
    hover_name='Country Code',
  )
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

- From this Pie chart we can observe that Switzerland is having highest percentage of GDP with 36.9% and india is having lowest percentage of GDP with 3.6%.
